```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
''' 
Finally, you have to  plot a graph with: 
○  X-axis: Number of concurrent clients (10, 20, ..., 100). 
○  Y-axis: Execution time (latency).
'''
```

```python
single_server = pd.read_csv('ser.csv')
multiThread_server = pd.read_csv('multiTh.csv')
multiProcess_server = pd.read_csv('multiPro.csv')
```

```python
single_server
```

```python
## Plotting the graph using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.title('Single Server')
plt.xlabel('Number of Concurrent Clients')
plt.ylabel('Execution Time (Latency)')

plt.plot(single_server['Clients'], single_server['Execution'], label='Single Threaded Server')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.title('MultiThread Server')
plt.xlabel('Number of Concurrent Clients')
plt.ylabel('Execution Time (Latency)')

plt.plot(multiThread_server['Clients'], multiThread_server['Execution'])
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))
plt.title('MultiProcess Server')
plt.xlabel('Number of Concurrent Clients')
plt.ylabel('Execution Time (Latency)')

plt.plot(multiProcess_server['Clients'], multiProcess_server['Execution'])
plt.show()
```
